### Excercise-1 ( 1 Mark)
John tracked his lunch bills for the past 5 days. He paid `$15.10, $12.05, $20.05, $17.75 and $16.25`. What was the average amount John paid per meal?  Print the result with 2 decimal places.      

```python
import numpy as np
lunch_bill = [15.10,12.05,20.05,17.75,16.25]
print("Average amount John paid was $",round(np.average(lunch_bill),2))
```

### Excercise-2 ( 1 Mark)

Mark received the following scores in his `Graphic Design` tests. His final grade is based on the mean of those scores. What will be Mark's final grade in that course? Print the result with 0 decimal places.

77, 84, 89, 78, 92


```python
scores = [77,84,89,78,92]
print("Mark's Final Grade in Graphic Design is",int(np.mean(scores)))
```

#### 4.1) Plot a histogram for the given below age of population with respect to bin.

| Population Age | 22,55,62,45,18,21,22,16,34,42,42,4,2,102,42,95,85,55,110,14,50,120,70,65,55,111,115,80,75,65,54,44,43,42,48  |

### Excercise-6 ( 1 Mark)
Using `linespace` method from `numpy` library create *200* sample data between the interval *1 and 50*. Plot the **normal distribution** using `matplotlib` library for the sample data.

```python
from scipy.stats import norm
import statistics

sample_data = np.linspace(start= 1,stop = 50, num = 200)
mean = statistics.mean(sample_data)
sd = statistics.stdev(sample_data)

plt.plot(sample_data,norm.pdf(sample_data,mean,sd))
plt.title("Normal Distribution using pdf")
plt.show()

plt.plot(sample_data,norm.cdf(sample_data,mean,sd))
plt.title("Normal Distribution using cdf")
plt.show()
```
